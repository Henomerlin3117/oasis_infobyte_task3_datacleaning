```python
#handling missing values
df['name'].fillna('Unknown', inplace=True)
df['host_name'].fillna('Unknown', inplace=True)
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(), inplace=True)
df['last_review'].fillna('2019-03-27', inplace=True)
df.isna().sum()
```

**STANDARDIZATION**

```python
# Check the unique values in the 'price' column
df['price'].nunique()
```

```python
# conversion rate from one currency to another (e.g., USD to INR)
conversion_rate = 83.37
```

```python
# Standardize the 'price' column by converting all values to price
df['price'] = df['price'] * conversion_rate
```

```python
# Print the standardized 'price' column
print("Standardized 'price' column:\n", df['price'])
```

```python
# Convert 'last_review' column to datetime
df['last_review'] = pd.to_datetime(df['last_review'], format='%Y-%m-%d')
```

```python
# standardizing last_review column
print("Standardized 'last_review' column:\n", df['last_review'])
```

**OUTLIER DETECTION**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot for price column
plt.figure(figsize=(6,4))
sns.boxplot(data=df['price'])
plt.show()
```

```python
from scipy import stats

z_scores = stats.zscore(df['price'])
abs_z_scores = np.abs(z_scores)
outliers_indices = (abs_z_scores > 3)
outliers = df[outliers_indices]
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
```

```python
# Histogram of price with outliers highlighted
plt.figure(figsize=(8,6))
plt.hist(df['price'], bins=30, alpha=0.5, color='blue', label='All Data')
plt.hist(outliers['price'], bins=30, alpha=0.5, color='red', label='Outliers')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Removed outliers from the dataset
data_cleaned = df[~outliers_indices]
```
